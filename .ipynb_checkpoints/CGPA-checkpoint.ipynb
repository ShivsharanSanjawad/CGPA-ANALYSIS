```python
def validate_sgpa(x):
    try:
        value = float(x)
        return value if 0 <= value <= 10 else np.nan
    except ValueError:
        return np.nan
```

```python
df.to_csv("resultant.csv")
df["SGPA"] = df["SGPA"].apply(validate_sgpa)
df["Total_GPC_Earned"] = pd.to_numeric(df["Total_GPC_Earned"], errors='coerce')
df["Total_Credits_Earned"] = pd.to_numeric(df["Total_Credits_Earned"], errors='coerce')
df.loc[df["Total_Credits_Earned"] == 0, "SGPA"] = 0
df["SGPA"] = df["SGPA"].fillna(df["Total_GPC_Earned"] /190)
df.to_csv("resultant.csv")
df.isnull().sum()
```

```python
columns_to_fill = [
    "EM_ESE", "EM_Total", "EM_Status", "EM_Grade", "EM_Credit", 
    "EM_GPA", "EM_GPC", "EG_ESE", "EG_Total", "EG_Status", 
    "EG_Grade", "EG_Credit", "EG_GPA", "EG_GPC"
]

# Fill NaN values in the specified columns with 0
df[columns_to_fill] = df[columns_to_fill].fillna(0)
df.isnull().sum()
```

```python
#used chatgpt for syntax
total_or_ese_cols = df.filter(like='Total').columns.tolist() + \
                    df.filter(like='ESE').columns.tolist() + \
                    df.filter(like='GPA').columns.tolist() + \
                    df.filter(like='GPC').columns.tolist() +\
                    df.filter(like='Credit').columns.tolist() +\
                    df.filter(like='UID').columns.tolist()
for col in total_or_ese_cols:
    df[col] = pd.to_numeric(df[col], errors='coerce')
df = df[df["SGPA"]!=0]
df.to_csv("resultant.csv")
```

```python
import pandas as pd
from pandas_profiling import ProfileReport
from pydantic_settings import BaseSettings

# Define your settings using Pydantic Settings
class Settings(BaseSettings):
    input_file: str = "FINAL.csv"  # Default value for the input file
    output_file: str = "CGPAreport.html"  # Default value for the output file

    class Config:
        env_file = ".env"  # Optional: Load environment variables from a .env file

# Initialize settings
settings = Settings()


# Create the profile report
profile = ProfileReport(df, title="Pandas Profiling Report", explorative=True)

# Save the report to an HTML file
profile.to_file(settings.output_file)  # Use output file from settings

print(f"Profile report generated successfully: {settings.output_file}")
```
